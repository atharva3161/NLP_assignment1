```
import nltk
import string
import numpy as np

def zipf_mean(x, c):
  return c*(np.sqrt(x))

def zipf_length(x, c):
  return (c*(1/x))

def heaps(x, k, b):
  return (k * np.power(x,b))
```

```
import pandas as pd
dataset = pd.read_csv("./tweets-dataset.csv")

from nltk.tokenize import TweetTokenizer
tokenizer = TweetTokenizer(strip_handles = 2)
```
